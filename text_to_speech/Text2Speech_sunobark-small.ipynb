```python
from transformers import pipeline
```

```python
text = "Hi My Name is Abubakar, I am Computer Engineer."

pipe = pipeline("text-to-speech", model="suno/bark-small", device="cuda")

output = pipe(text)
```

```python
output # it will be a numpy array
```

```python
#now decoding the numpy array to create the audio and listen it
from IPython.display import Audio

Audio(output["audio"], rate=output["sampling_rate"])
```
